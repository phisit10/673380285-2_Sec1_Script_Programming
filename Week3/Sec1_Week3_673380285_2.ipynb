Lab 3.1: โปรแกรมสร้างสูตรคูณ (Multiplication Table Generator)

```python
#Part 1: ตารางสูตรคูณแม่เดียว (Single Number Table)

n=int(input("Enter a number for its muitiplication table : "))
for i in range (1,11):
  print(f"{n} * {i} = {i*n}")
```

```python
#Part 2: ตารางสูตรคูณแบบเต็ม (Full Multiplication Grid) - โจทย์ท้าทาย/ส่วนขยาย

print("12*12 Multiplication Table: ")
for i in range (1,13):
  for j in range (1,13):
    print(f"{i*j:4}", end=' ')
  print()
```

Lab 3.2: โปรแกรมการนับถอยหลัง & เกมทายตัวเลข (Countdown & User Guessing Game)

```python
#Part 1: โปรแกรมตัวนับถอยหลัง (Countdown Timer)

start = int(input("Enter the Starting number for countdown: "))
while start >= 0:
  print(f"{start}")
  start = start - 1
print("Blast off!")
```

```python
#Part 2: เกมทายตัวเลขอย่างง่าย (Simple Guessing Game):

secert_number = 50

while True:
  guess = int(input("Guess the secret number (between 1 and 100): "))

  if guess > secert_number:
    print(f"{guess} is Too high! Try again")
  elif guess < secert_number:
    print(f"{guess} is Too low! Try again")
  else:
    print("Congratulations! You guessed it")
    break
```

Activity 3.2: Pattern Printing with Nested Loops

```python
for i in range(1,6):
  for j in range(1,6):
    print("*", end='')
  print()
```

```python
for i in range (0,6):
  for j in range(i) :
    print("*",end=' ')
  print()
```
