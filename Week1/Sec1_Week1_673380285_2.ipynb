# คำนวณพื้นที่รูปเลขาคณิตพื้นฐานอย่างง่าย

```python
print("Area of shapes")

print("1.Regtangle")
print("2.Circle")
print("3.Triangle")

Check = int(input("Enter a number: "))

print(" ")
if Check == 1:
    print("Find the Area of Regtangle")
    Wid = float(input("Enter a width : "))
    Hid = float(input("Enter a height : "))
    A = Wid * Hid
    print("Area is", A)
elif Check == 2:
    print("Find the Area of Circle")
    R = float(input("Enter a Radius : "))
    A = 3.14 * R * R
    print("Area is", A)
elif Check == 3:
    print("Find the Area of Triangle")
    B = float(input("Enter a Base : "))
    H = float(input("Enter a height : "))
    A = 0.5 * B * H
    print("Area is", A)

else:
    print("Invalid choice")


```
